```python
from dotenv import load_dotenv

load_dotenv()
```

```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate, FewShotChatMessagePromptTemplate

chat = ChatOpenAI(model="gpt-3.5-turbo", temperature=0.1)
```

```python
examples = [
    {
        "movie": "Inception",
        "answer": """
Movie: Inception
Director: Christopher Nolan
Main Cast: Leonardo DiCaprio, Joseph Gordon-Levitt, Elliot Page, Tom Hardy
Genre: Science Fiction, Action, Thriller
Synopsis: A skilled thief enters people's dreams to steal secrets. He is offered a chance to have his criminal record erased if he can plant an idea inside someone's mind.
""",
    },
    {
        "movie": "The Godfather",
        "answer": """
Movie: The Godfather
Director: Francis Ford Coppola
Main Cast: Marlon Brando, Al Pacino, James Caan, Robert Duvall
Genre: Crime, Drama
Synopsis: The aging head of a powerful mafia family transfers control of his empire to his reluctant son, leading to conflict, loyalty tests, and tragedy.
""",
    },
    {
        "movie": "Parasite",
        "answer": """
Movie: Parasite
Director: Bong Joon-ho
Main Cast: Song Kang-ho, Lee Sun-kyun, Cho Yeo-jeong, Choi Woo-shik, Park So-dam
Genre: Thriller, Drama, Dark Comedy
Synopsis: A poor family schemes to become employed by a wealthy household, but their plan leads to unexpected consequences and a sharp conflict between social classes.
""",
    },
]
```

```python
example_prompt = ChatPromptTemplate.from_messages(
    [
        ("human", "Tell me about the movie {movie}"),
        ("system", "{answer}"),
    ]
)
```

```python
fewshot_prompt = FewShotChatMessagePromptTemplate(
    examples=examples,
    example_prompt=example_prompt,
    # suffix="Tell me about the movie {movie}",
    # input_variables=["movie"],
)
```

```python
final_prompt = ChatPromptTemplate.from_messages(
    [
        ("system", """
        You are a movie expert.
        You must always answer using exactly the same format as the examples.
        Do not add extra sections.
        """),
        fewshot_prompt,
        ("human", "Tell me about the movie {movie}"),
    ]
)

chain = final_prompt | chat
```
